```python
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
```

```python
# Loading the dataset
X_test = pd.read_csv('X_test')
X_test_copy = X_test.copy()
Y_test = pd.read_csv('Y_test')
```

```python
# Loading the necessary encoders, scalars, model
with open('label_encoder_sex.pkl','rb') as file:
    label_encoder_sex = pickle.load(file)

with open('label_encoder_smoke.pkl','rb') as file:
    label_encoder_smoke = pickle.load(file)

with open('onehot_region.pkl','rb') as file:
    onehot_region = pickle.load(file)    

with open('scalar.pkl','rb') as file:
    scalar = pickle.load(file)    

model = load_model('model.h5')   
```

```python
# Label Encoding
X_test['sex'] = label_encoder_sex.transform(X_test['sex'])
X_test['smoker'] = label_encoder_smoke.transform(X_test['smoker'])
```

```python
# One Hot Encoding
X_test_region = onehot_region.transform(X_test[['region']])
col_names = onehot_region.get_feature_names_out()
encoded_df_test = pd.DataFrame(X_test_region,columns = col_names )
X_test = X_test.drop('region',axis = 1)
X_test = pd.concat([X_test, encoded_df_test], axis = 1)
```

```python
# Scaling
X_test = scalar.transform(X_test)
```

```python
# Prediction
predict = model.predict(X_test)
predict
```

```python
predictions_df = pd.DataFrame(predict, columns=['Predicted_Y'])
result_df = pd.concat([X_test_copy.reset_index(drop=True), Y_test.reset_index(drop=True), predictions_df], axis=1)
result_df.rename(columns={'charges': 'Original Y'}, inplace=True)
```
